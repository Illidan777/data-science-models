```python
# Next, we have a problem in that our target is syntactically difficult to build a model. Since one player can play in several positions, we need to 
# create copies of the player for each of his positions that he can occupy
separated_df = df.assign(player_positions=df['player_positions'].str.split(', ')).explode('player_positions')
separated_df
```

```python
# It is also necessary to encode the features that we will use in the model
from sklearn.preprocessing import LabelEncoder

player_position_encoder = LabelEncoder()

separated_df['club_position_encoded'] = player_position_encoder.fit_transform(separated_df['club_position'])
separated_df['player_positions_encoded'] = player_position_encoder.transform(separated_df['player_positions'])
separated_df
```

```python
# Here I analyzed how unbalanced the data is. And I realized that this needs to be fixed. However, we will build models on what we have already processed
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight('balanced', classes=separated_df['player_positions_encoded'].unique(), y=separated_df['player_positions_encoded'])
print(class_weights)
```

```python
#Let's write a custom function that will evaluate the accuracy of the model. Because the main function of the model is to predict at least one position in 
# which a football player can play

def custom_accuracy_score(encoder, x_test, _pred, data, classifier_title):
    inverse_transformed_pred = encoder.inverse_transform(_pred)
    correct_predictions = 0
    pred_series = pd.Series(inverse_transformed_pred, index=x_test.index)
    for index, value in pred_series.items():
        original_record = data.loc[index]
        print(f"Index: {index}, Predicted value: {value}, Original value: {original_record['player_positions']}\n")
        if value in original_record['player_positions']:
            correct_predictions += 1
    print(f'Custom accuracy({classifier_title}):', (correct_predictions / len(inverse_transformed_pred)) * 100)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# We will select only those columns that most influence our target and which describe only the psychological, physical and professional characteristics of 
# the player, as I described above. Let's divide our dataset into test and train splits.

# Let's test several models of clasiffication. (Logistic regression, KNeighborsClassifier, GaussianNB, Neural Network). Logistic regression gave us the 
# best result.
# I also tested the model with and without club_position. And using the model without club_position gave the best result. But unfortunately, I did not 
# have time to study the nature of this behavior.

X = separated_df.drop(
    columns=['long_name', 'overall', 'potential', 'value_eur', 'wage_eur', 'club_position', 'club_team_id', 'club_name',
             'league_name', 'league_level', 'nationality_name', 'player_positions', 'player_positions_encoded',
             'club_position_encoded'])
Y = separated_df.loc[:, ['player_positions_encoded']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=101)

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=101, solver='sag', multi_class='auto')
classifier.fit(X_train, Y_train)

pred = classifier.predict(X_test)

accuracy = accuracy_score(Y_test, pred)

custom_accuracy_score(player_position_encoder, X_test, pred,  df, 'Logistic regression')
print('General accuracy(Logistic regression): ', accuracy * 100)
```

```python
from sklearn import neighbors

#Методом підбору кількість сусідів 8 показало накрайщи результат, варіація була приблизно в 1 два відсотка
knn_model = neighbors.KNeighborsClassifier(n_neighbors=25)
knn_model.fit(X_train, Y_train)

pred = knn_model.predict(X_test)

accuracy = accuracy_score(Y_test, pred)

custom_accuracy_score(player_position_encoder, X_test, pred,  df, 'KNeighborsClassifier')
print('General accuracy(KNeighborsClassifier): ', accuracy * 100)
```

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(X_train, Y_train)

pred = classifier.predict(X_test)

accuracy = accuracy_score(Y_test, pred)

custom_accuracy_score(player_position_encoder, X_test, pred,  df, 'GaussianNB')
print('General accuracy(GaussianNB): ', accuracy * 100)
```

```python
from tensorflow import keras
from keras.layers import LeakyReLU

model = keras.Sequential()

# I used the relay face activation function using a life hack)) because our output values ​​can be more than 1. The number of lefers was selected by sampling and
# of errors
input_layer = keras.layers.Dense(3, input_shape=[X_train.shape[1]], activation=LeakyReLU())
model.add(input_layer)

hidden_layer_1 = keras.layers.Dense(16, activation=LeakyReLU())
model.add(hidden_layer_1)

hidden_layer_2 = keras.layers.Dense(32, activation=LeakyReLU())
model.add(hidden_layer_2)

hidden_layer_3 = keras.layers.Dense(64, activation=LeakyReLU())
model.add(hidden_layer_3)

hidden_layer_4 = keras.layers.Dense(128, activation=LeakyReLU())
model.add(hidden_layer_4)

output_layer = keras.layers.Dense(100, activation='softmax')
model.add(output_layer)

# Loss function sparse_categorical_crossentropy because this function requires int numeric classes. I could use categorical_crossentropy, but it needs it
# one hot encoding, and we don't need it here because we already have classes in numerical format. And for example such functions as mean_squared_error or mean_absolute_error they are
# are aimed at solving the regression problem, which is not suitable in our case. Well, the binary_crossentropy function does not suit us because we do not have a binary result
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, Y_train,
            
                    epochs=200,
                    steps_per_epoch=10,
                    batch_size=30)

loss, accuracy = model.evaluate(X_test, Y_test)

custom_accuracy_score(player_position_encoder, X_test, np.argmax(model.predict(X_test), axis=1),  df, 'Neural Network')
print('General accuracy(Neural Network): ', accuracy * 100)
```
